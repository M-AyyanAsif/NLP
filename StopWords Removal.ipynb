```python
nlp= spacy.load('en_core_web_sm')
doc= nlp('We just opened our wings, the flying part is coming soon')
```

```python
for token in doc:
    if not token.is_stop:
        print(token)
```

```python
def process(text):
    doc= nlp(text)
    
    stop_word_removal= [token.text for token in doc if not token.is_stop]
    return " ".join(stop_word_removal)
```

```python
process("Musk wants time to prepare for a trial over his")
```

```python
process("Heloo! hope are you i am fine hope you are also very good and fit andf enjoying your life")
```
