```python
import nltk
nltk.download('punkt')
```

```python
text="Hello! how are you. I am good . Where are you now days.I am waiting for you many day(specific Person)"
```

```python
sent = nltk.sent_tokenize(text)
print(sent)
```

```python
for word in sent:
    words= nltk.word_tokenize(word)
    print(words)
```
