```python
data = [
    ("What is machine learning?",
     "Machine learning is a method where computers learn patterns from data."),

    ("What is artificial intelligence?",
     "Artificial intelligence is the simulation of human intelligence by machines."),

    ("What is data science?",
     "Data science involves extracting insights from data using statistics and algorithms."),

    ("What is supervised learning?",
     "Supervised learning uses labeled data to train models."),

    ("What is unsupervised learning?",
     "Unsupervised learning finds patterns in unlabeled data."),

    ("What is a dataset?",
     "A dataset is a collection of structured or unstructured data."),

    ("What is overfitting?",
     "Overfitting happens when a model learns noise instead of patterns."),

    ("What is accuracy in machine learning?",
     "Accuracy measures how many predictions are correct."),

    ("What is feature extraction?",
     "Feature extraction converts raw data into meaningful features."),

    ("What is a model in machine learning?",
     "A model is a mathematical representation learned from data.")
]
```

```python
import pandas as pd
df= pd.DataFrame(data, columns=["Questions","Answer"])
df['label']=df.index
```

```python
df.head(4)
```

```python
df.shape
```

```python
import re

def preprocess(text):
    text= text.lower()
    text= re.sub(r"[^a-z\s]", "", text)
    text= re.sub(r"\s+", " ", text).strip()
    return text
```

```python
df["Cleaned_Questions"]= df["Questions"].apply(preprocess)
```

```python
df.head(4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(df['Cleaned_Questions'],df['label'],test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer(
    ngram_range=(1,3),
    stop_words="english"
)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test) 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=500)
model.fit(X_train_vec,y_train)
y_pred=model.predict(X_test_vec)
```

```python

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
def answer_questions(question):
    question = preprocess(question)
    vector = vectorizer.transform(["question"])
    label = model.predict(vector)[0]
    return df.loc[label, "Answer"]
    
```

```python
query = "Can you explain what overfitting means?"
response = answer_questions(query)

print("Q:", query)
print("A:", response)
```
