```python
text = """Kiran want to know the famous foods in each state of India. So, he opened Google and search for this question. Google showed that
in Delhi it is Chaat, in Gujarat it is Dal Dhokli, in Tamilnadu it is Pongal, in Andhrapradesh it is Biryani, in Assam it is Papaya Khar,
in Bihar it is Litti Chowkha and so on for all other states"""
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
all_gpe_names=[]
for ent in doc.ents:
    if ent.label_ == 'GPE':
        all_gpe_names.append(ent)
```

```python
print("ALl cities are:",all_gpe_names)
print("Lenght are:",len(all_gpe_names))
```

```python
text2 = """Sachin Tendulkar was born on 24 April 1973, Virat Kholi was born on 5 November 1988, Dhoni was born on 7 July 1981
and finally Ricky ponting was born on 19 December 1974."""
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text2)
```

```python
all_birth_date= []
for ent in doc.ents:
    if ent.label_ == 'DATE':
        all_birth_date.append(ent)
```

```python
print("Birth Dates are:",all_birth_date)
print(len(all_birth_date))
```
