```python
import spacy
```

```python
nlp= spacy.blank('en')
```

```python
text= "Captain america ate 100$ of samosa. Then he said I can do this all day."
doc= nlp(text)
```

```python
for sent in doc:
    print(sent)
```

```python
doc= nlp(text)
```

```python
for token in doc:
    print(token, " | ", spacy.explain(token.pos_), " | ",token.lemma_)
```

```python
source_nlp= spacy.load('en_core_web_sm')

nlp=spacy.blank('en')
nlp.add_pipe("ner", source=source_nlp)
nlp.pipe_names
```

```python
p_nlp= spacy.load('en_core_web_sm')
n_nlp= spacy.blank('en')

n_nlp.add_pipe("tok2vec", source=p_nlp)
n_nlp.pipe_names
```
